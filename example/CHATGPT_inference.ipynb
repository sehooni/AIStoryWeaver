# ChatGPT API 기본

```python
import openai

# 발급받은 API 키 설정
OPENAI_API_KEY = "..."

# openai API 키 인증
openai.api_key = OPENAI_API_KEY
```

```python
# 모델 - GPT 3.5 Turbo 선택
model = "gpt-3.5-turbo"

# 질문 작성하기
query = "액션 RPG 게임을 만들거야! 스토리를 만들어줘"

# 메시지 설정하기
messages = [{
    "role": "system",
    "content": "You are a helpful assistant."
}, {
    "role": "user",
    "content": query
}]
```

```python
# ChatGPT API 호출하기
response = openai.ChatCompletion.create(model=model, messages=messages)
answer = response['choices'][0]['message']['content']
answer
```

# ChatGPT finetune model use

```python
from openai import OpenAI
api_key = "..."
client = OpenAI(api_key=api_key)

completion = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "system", "content": "너는 창의적인 게임 퀘스트 스토리 제작자야."},
    {"role": "user", "content": "너는 퀘스트 제목, 발단, 전개, 위기, 절정, 결말로 구성된 스토리를 만들어줄거야. 등장인물은 모험가, 지그문트, 라하르트, 에아달린, 아델, 가름, 에버그레이스 야. 주요 이벤트는 모험가가 지그문트의 계획을 저지하고 마력핵의 폭주를 막은 후, 에아달린의 쇄빙선을 타고 슈사이어로 향하는 이야기야"}
  ]
)
```
